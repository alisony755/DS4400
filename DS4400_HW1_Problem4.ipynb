```python
# Install numpy and sympy
import numpy as np
import sympy as sp
```

```python
# Generate at random 3 matrices of size 3 x 3 and fill each entry with a random integer chosen from -10 to 10
matrix_a = np.random.randint(-10, 11, size=(3, 3))
matrix_b = np.random.randint(-10, 11, size=(3, 3))
matrix_c = np.random.randint(-10, 11, size=(3, 3))

# Convert matrices to SymPy matrices (to be able to compute exact inverse)
sympy_a = sp.Matrix(matrix_a.tolist())
sympy_b = sp.Matrix(matrix_b.tolist())
sympy_c = sp.Matrix(matrix_c.tolist())

# Print matrices
print(sympy_a)
print()
print(sympy_b)
print()
print(sympy_c)
```

```python
# Get inverse of matrix A
inverse_a = sympy_a.inv()
print(inverse_a)
```

```python
# Get inverse of matrix B
inverse_b = sympy_b.inv()
print(inverse_b)
```

```python
# Get inverse of matrix C
inverse_c = sympy_c.inv()
print(inverse_c)
```

```python
# Check if matrix A * inverse of matrix A = identity matrix
identity_a = matrix_a @ inverse_a
print(identity_a)
```

```python
# Check if matrix B * inverse of matrix B = identity matrix
identity_b = matrix_b @ inverse_b
print(identity_b)
```

```python
# Check if matrix C * inverse of matrix C = identity matrix
identity_c = matrix_c @ inverse_c
print(identity_c)
```
