### 5.1

1. For each feature, write code to compute the average value, the min and max values, as well as its variance.
    
2. Which features have the lowest and the highest average? Include the feature name and their average values for the features with the lowest and highest average.
> **long has the lowest average value (-122.21389640494147) and price has the highest average value (540088.1417665294)**


    
3. Which features have the lowest and the highest variance? Include the feature name and their variance values for the features with the lowest and highest variance.
> **waterfront has the lowest variance (0.007485225502689098) and price has the highest variance (134782378397.24681)**

```python
# Compute the average value, the min and max values, and variance for each feature
for column_name, column_data in df_house.items():
    print(f"{column_name} stats:")

    # Calculate the average value for a feature
    feature_mean = df_house[f"{column_name}"].mean()

    # Print the average value of the feature
    print(f"{column_name} average value: {feature_mean}")

    # Calculate the min value for a feature
    feature_min = df_house[f"{column_name}"].min()

    # Print the min value of the feature
    print(f"{column_name} min value: {feature_min}")

    # Calculate the max value for a feature
    feature_max = df_house[f"{column_name}"].max()

    # Print the max value of the feature
    print(f"{column_name} max value: {feature_max}")

    # Calculate the variance for a feature
    feature_var = df_house[f"{column_name}"].var()

    # Print the variance of the feature
    print(f"{column_name} variance: {feature_var}")
    print('\n')
```

1. Compute the **correlation coefficient** of each feature with the response. **Include a table** with the correlation coefficient of each feature with the response.

2. Which features are positively correlated (i.e., have positive correlation coefficient) with the response?

3. Which feature has the **highest positive correlation** with the response?

```python
# Compute correlation matrix
corr_matrix = df_house.corr(numeric_only =True)

# Get correlations with response variable (price)
price_corr = corr_matrix['price'].drop('price').sort_values(ascending=False)

# Display correlation coefficient of each feature as a table
price_corr.to_frame(name = "Correlation with Price")
```
